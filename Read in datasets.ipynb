```python
#READ IN LIBRARIES/PACKAGES

import os
import shutil
import pandas as pd
from datetime import datetime
from lxml import objectify
import sys

```

```python
#USER INPUT: choose location and file for input *.csv

#inputloc = (r"\\Data\input_datasets\")
#outputloc = (r"\\Data\output_datasets\")
filetype = ('csv')

inputloc = input("Enter input file location:" )
inputloc = inputloc +"\\"     #+ r"\"
print("Input file location: " + inputloc)
             
#outputloc = input ("Enter output file lcoation:")
#outputloc = outputloc+"\\"    #+ r"\"
#print("Output file lcoation: " + outputloc)


#USER INPUT FILENAME

inputfile = input("Enter filename: ")
inputfile = inputfile + ".csv"
print("Input filename is: " + inputfile)


#INPUT FILE PATH
inputpath = inputloc + inputfile
print ("Full input filepath:" + inputpath)

if not os.path.exists(inputpath):
    print('{} not found'.format(inputpath))
    sys.exit()
```

```python
#READ IN AS DATAFRAME

df = pd.read_csv(inputpath)

df.head()
```

```python
#SUMMARY DATA FOR THE DATASET


#CHECK DATA HEADINGS

print ("HEAD", df.head(), '\n')



#DESCRIBE THE DATA

print('DESCRIBE', df.describe(), '\n')



#SHAPE OF THE DATA: number of columns and rows

print ("SHAPE", df.shape, '\n')



#INFO: infomormation of datatypes

print ("INFO", df.info(), '\n')
```

```python
#FORMAT HEADERS TO UPPERCASE
df.columns = map(str.upper, df.columns)
df.head()
```

```python
#FORMATTING

#EVENT TYPE inconsistenses with case
df['EVENT_TYPE'] = df['EVENT_TYPE'].replace({'Strategic Developments': 'Strategic developments', 
                                            'Explosion/Remote Violence':'Explosions/Remote violence', 
                                            'Explosions/remote violence':'Explosions/Remote violence'})

df['SUB_EVENT_TYPE'] = df['SUB_EVENT_TYPE'].replace({'Peaceful Protest':'Peaceful protest', 
                                                    'Armed Clash':'Armed clash', 'Mob Violence':
                                                     'Mob violence'})


#STANDARDIZE EVENT FIELDS
#strip white spaces
df["EVENT_TYPE"]=df["EVENT_TYPE"].str.strip()
df["SUB_EVENT_TYPE"]=df["SUB_EVENT_TYPE"].str.strip()
df["EVENT_ID_CNTY"]=df["EVENT_ID_CNTY"].str.strip()
```

```python
#USER OUTPUT: Choose location and filname for *.csv output file

outputloc = input ("Enter output file lcoation:")
outputloc = outputloc+"\\"    #+ r"\"
print("Output file location: " + outputloc)

outputfile = input("Enter desired output filename: ")
outputfile = outputfile + ".csv"
print("Output filename is: " + outputfile)


#OUTPUT FILE PATH
outputpath = outputloc + outputfile
print ("Full output filepath:" + outputpath)
```

```python
#CREATE OUTPUT FILE: will be formatted and can be used for back-up
df.to_csv(outputpath, index = False)

print ("Output *.csv Complete")
```
